# African Climate Cross-Country Comparison

This notebook synthesizes the cleaned datasets from Ethiopia, Kenya, Sudan, Tanzania, and Nigeria to identify relative climate vulnerability and produce data-driven insights for COP32.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import glob
import os

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (14, 8)
```

## 1. Load and Concatenate Cleaned Data

```python
clean_files = glob.glob("../data/*_clean.csv")
dfs = [pd.read_csv(f) for f in clean_files]

if dfs:
    df_all = pd.concat(dfs, ignore_index=True)
    df_all['Date'] = pd.to_datetime(df_all['Date'])
    print("Combined DataFrame shape:", df_all.shape)
    display(df_all['Country'].value_counts())
else:
    print("No clean CSV files found in data/ directory.")
```

## 2. Temperature Trend Comparison

```python
if 'df_all' in locals():
    # Monthly average T2M per country
    monthly_compare = df_all.groupby(['Country', pd.Grouper(key='Date', freq='M')])['T2M'].mean().reset_index()
    
    plt.figure(figsize=(15, 7))
    sns.lineplot(data=monthly_compare, x='Date', y='T2M', hue='Country')
    plt.title('Monthly Avg. Temperature Comparison (2015-2026)')
    plt.ylabel('Temperature (°C)')
    plt.show()
```

## 3. Precipitation Variability Comparison

```python
if 'df_all' in locals():
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df_all, x='Country', y='PRECTOTCORR')
    plt.title('Daily Precipitation Distribution Comparison')
    plt.ylabel('Precipitation (mm/day)')
    plt.yscale('log') # Log scale to handle extremes
    plt.show()
```

## 4. Extreme Event Frequency

```python
if 'df_all' in locals():
    # Extreme Heat Days (> 35 °C)
    extreme_heat = df_all[df_all['T2M_MAX'] > 35].groupby(['Country', 'YEAR']).size().reset_index(name='Heat_Days')
    
    plt.figure(figsize=(12, 6))
    sns.barplot(data=extreme_heat, x='YEAR', y='Heat_Days', hue='Country')
    plt.title('Annual Count of Extreme Heat Days (>35°C)')
    plt.show()
```
